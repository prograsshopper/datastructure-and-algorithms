```python
class Node(object):
    def __init__(self, key=None, next=None):
        self.key = key
        self.next = next

class SingleLinkedList(object):
    def __init__(self):
        self.head = Node("head", None)
    
    def push_front(self, key):
        try: 
            new_node = Node(key, None)
            if not self.head.next:
                self.head.next = new_node
                return True
            new_node.next = self.head.next
            self.head.next = new_node
            return True
        except Exception as e:
            print(e)
            return False
        
    def top_front(self):
        return self.head.next
        
    def pop_front(self):
        pop_node = self.head.next
        self.head.next = self.head.next.next
        return pop_node
    
    def push_back(self, key):
        try:
            new_node = Node(key, None)
            current = self.head
            while current:
                if not current.next:
                    current.next = new_node
                    break
                current = current.next
        except Exception as e:
            raise e

    def top_back(self):
        current = self.head
        while current:
            if not current.next:
                break
            current = current.next
        return current
        
    def pop_back(self):
        try:
            target = None
            current = self.head
            while True:
                if not current.next.next:
                    target = current.next
                    current.next = None
                    break
                current = current.next
            return target
        except Exception as e:
            print(e)
            return False

    def find(self, key):
        current = self.head
        while current:
            if current.key == key:
                find = True
                return current
            if not current.next:
                break
            current = current.next
        return False
            
    def erase(self, key):
        current = self.head
        while current.next:
            if current.next.key == key:
                current.next = current.next.next
                break
            current = current.next    
        return True
    
    def empty(self):
        if self.head.next:
            return False
        return True
    
    def add_before(self, target_node, key):
        new_node = Node(key, None)
        current = self.head
        while current:
            if current.next == target_node:
                new_node.next = current.next
                current.next = new_node
                break
            current = current.next
    
    def add_after(self, target_node, key):
        new_node = Node(key, None)
        new_node.next = target_node.next
        target_node.next = new_node
    
    def print_all_items(self):
        current = self.head
        while current:
            print(current.key)
            current = current.next
```

```python
single_list = SingleLinkedList()
single_list.push_front(1)
print(single_list.top_front().key)
single_list.push_back(2)
print(single_list.top_back().key)

single_list.push_back(3)
single_list.push_back(4)
single_list.push_back(5)

target = single_list.find(3)
print(target.key)
print(single_list.find(10))
print(single_list.empty())

single_list.add_before(target, 11)
single_list.add_after(target, 15)

single_list.print_all_items()
```
